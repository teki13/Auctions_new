```python
#Import relevant libraries
import pysurveycto
import pandas as pd
import time
import datetime
```

```python
#connect to the surveyCTO API
scto = pysurveycto.SurveyCTOObject('ghananatlsurvey','tr1328@nyu.edu', 'Ping10gratis!')
```

```python
#specify date
date_input = datetime.datetime(2021,8,10)

#retrive the form data in a json format
data = scto.get_form_data('negotiation_data_survey', format='json', oldest_completion_date = date_input)

#Convert the json folder to a pandas dataframe
df = pd.json_normalize(data)
```
